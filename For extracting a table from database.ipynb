# For extracting a table from database

```python
import pandas as pd
import sqlite3

# Connect to your database
conn = sqlite3.connect('inventory.db')

# Load the table into a DataFrame
vendor_sales_summery = pd.read_sql("SELECT * FROM vendor_sales_summary", conn)

# Export to CSV
vendor_sales_summery.to_csv("vendor_sales_summary.csv", index=False)
```
